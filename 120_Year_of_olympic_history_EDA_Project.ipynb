find and display the count of medals (bronze silver gold )  won in each city and present the result in a table format

```python
data.groupby(['City','Medal']).size().unstack()
```

sort the medal counts per city based on the total number of medals (bronze silver gold)

```python
medal=data.groupby(['City','Medal']).size().unstack()
```

```python
medal['Total']=medal[['Bronze','Silver','Gold']].sum(axis=1)
```

```python
medal.sort_values(by='Total',ascending=False)
```
